```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
!wget https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv
```

```python
df = pd.read_csv("titanic.csv")
df.head(10)
```

seleccion de features

```python
df = df.drop(columns=["PassengerId", "Name", "Ticket","Cabin"])
```

```python
df.head(10)
```

```python
df.isna().any()
```

```python
df.isna().sum()
```

```python
df.count()
```

```python
df.Age = df.Age.fillna(0)
df.head(10)
```

```python
#df.Age.mean = (df.Age.mean())
#df.head(10)
```

```python
df.Embarked = df.Embarked.fillna("Other")
df.head()
```

```python
fig = plt.figure(figsize =(10,10))
ax = fig.gca()
df.hist(ax=ax)
```

Notacion one-hot

```python
df_onehot = pd.get_dummies(df)
df_onehot.head(10)
```

Evaluacion de modelos de Machine Learning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score,roc_curve, roc_auc_score
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_onehot.iloc[:, 1:], df_onehot.iloc[:,0], test_size=0.3, shuffle =True, random_state=42)
```

```python
X_train[:10]
```

```python
X_test[:10]
```

```python
y_train[:10]
```

```python
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
probas_rl = clf.predict_proba(X_test)
probas_rl[:10]
```

```python
prob_rl = probas_rl[:,1]
prob_rl[:10]
```

```python
fpr_rl ,tpr_rl, thresholds_rl = roc_curve(y_test, prob_rl)
```

```python
plt.figure(figsize=(10, 10))
plt.title('Curva ROC')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.plot(fpr_rl, tpr_rl, marker='.', markersize=10)
plt.plot([0,1], [0,1], linestyle = "--")
```

```python
roc_auc_score(y_test, prob_rl)
```

```python
accuracy_score(y_test, y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
import seaborn as sns
```

```python
sns.heatmap(cm, annot=True, fmt="g")
```

```python
precision_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(max_depth=10)
rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```
